# Prédire la valeur d'un bien immobilier en utilisant la régression linéaire simple. Exemple du Boston Housing Dataset

```python
import sklearn as sk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
# Chargement des données du Boston Housing Dataset
from sklearn.datasets import load_boston
boston_dataset = load_boston()
print(boston_dataset.keys())
```

```python
# Afficher la description des données
print(boston_dataset.DESCR)
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston.head()
```

```python
# Ajout de la colonne MEDV
boston['MEDV'] = boston_dataset.target
boston.head()
```

```python
# Check des indices de corrélation entre les différentes variable grace à une heatmap(1 = high correlation and -1 = negative correlation)
plt.figure(figsize=(10, 8))
correlation_matrix = boston.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True, vmin=-1.0, vmax=1.0)
plt.show()
```

```python
# utilisation d'un scatterplot pour voir comment les 13 features evoluent avec MEDV
plt.figure(figsize=(60, 10))

features = ['CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
target = boston['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title("Variation in House prices")
    plt.xlabel(col)
    plt.ylabel('"House prices in $1000"')
```

```python
X = boston.drop(['MEDV'], axis=1)
X.to_numpy()
y.to_numpy()
print(X.shape)
print(y.shape)
```

```python
# séparation en training(80%) et testing(20%) sets. Evaluation de la performance du modèle sur des données non vues.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Afficher la distribution des différents jeux de données (nombre d'exemples dans le jeu de données initial, dans le jeu de train et dans le jeu de test en fonction des valeurs de MDEV) sous forme d'histogrammes
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.histplot(y, bins=30, kde=True, stat="density", linewidth=0, color='b')
sns.histplot(y_train, bins=30, kde=True, stat="density", linewidth=0, color='g')
sns.histplot(y_test, bins=30, kde=True, stat="density", linewidth=0, color ='r')
plt.show()
```

```python
# Training et testing du model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
# fit = entrainement sur le model, donc ne jamais le faire sur les données test. A faire seulement sur les données train.
```

```python
# Evaluation du model pour le training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = r2_score(y_train, y_train_predict)

print("Performance du training set :")
print('RMSE : {}'.format(rmse))
print('R2 score : {}'.format(r2))
print("\n")

# Evaluation du model pour le testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2 = r2_score(y_test, y_test_predict)

print("Performance du testing set :")
print('RMSE : {}'.format(rmse))
print('R2 score : {}'.format(r2))
```

```python
plt.figure(figsize=(10,8))
plt.plot(y_test, y_test_predict, 'o')
plt.plot(y_test, y_test, label='y = X')
plt.xlabel('y_test reel')
plt.ylabel('y_test predict')
plt.legend()
plt.show()
```
